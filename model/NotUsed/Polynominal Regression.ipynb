```python

import numpy as np
import matplotlib.pyplot as plt  # To visualize
import pandas as pd  # To read data
# Plot style.
plt.style.use("ggplot")

# Plot size.
plt.rcParams['figure.figsize'] = [14, 8]
```

```python
data = pd.read_csv('powerproduction.csv')  # load data set
```

```python
X = data[['speed']]
y = data[['power']]
```

```python
#plt.plot(X, y , 'x',markerfacecolor='blue', markersize=5)
# plt.xlabel("Power")
# plt.ylabel("Speed")

```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

pre_process = PolynomialFeatures(degree = 4)
# Setting the PolynomialFeatures degrees
X_Poly = pre_process.fit_transform(X)
X_Poly
```

```python
pr_model = LinearRegression()
# pr_model.fit(X,y, epochs=500, batch_size=10)
y_pred = pr_model.predict(X_Poly)

y_pred
```

```python
#plt.scatter(X, y ,c = "blue", label='actual')
plt.plot(X, y,'o', label='actual',markerfacecolor='blue', markersize=3)
plt.xlabel("Power")
plt.ylabel("Speed")
plt.plot(X,y_pred,c = "orange", label='prediction')
```
