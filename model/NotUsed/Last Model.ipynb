```python

import numpy as np
import matplotlib.pyplot as plt  # To visualize
import pandas as pd  # To read data
from sklearn.linear_model import LinearRegression
import tensorflow as tf

# Plot style.
plt.style.use("ggplot")

# Plot size.
plt.rcParams['figure.figsize'] = [14, 8]
```

```python
data = pd.read_csv('powerproduction.csv')  # load data set
```

```python
data = pd.DataFrame(data,columns=['speed','power'])
data
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])
])
model.compile(loss='mean_squared_error', 
              optimizer=tf.keras.optimizers.Adam(0.1))
```

```python
history = model.fit(data.power, data.speed, epochs=500, batch_size=10)
```

```python
plt.xlabel("Epoch Number")
plt.ylabel("Loss Magnidute")
plt.plot(data.speed, data.power,'o', label='actual',markerfacecolor='blue', markersize=3)
plt.xlabel("Power")
plt.ylabel("Speed")
plt.plot(data.speed,history, )
```
